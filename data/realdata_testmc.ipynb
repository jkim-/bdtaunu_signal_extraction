```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
mc = np.genfromtxt('mc.central/test.csv')
data = np.genfromtxt('data.central/data.csv')
```

```python
print mc.shape, data.shape
```

```python
nbins=300
```

```python
fig = plt.figure(figsize=(10,16))
ax = fig.add_subplot(211)
ax.hist(mc[:,0], normed=True, weights=mc[:,2], bins=nbins, histtype='step', label='MC');
ax.hist(data[:,0], normed=True, weights=data[:,2], bins=nbins, histtype='step', label='Data');
ax.set_xlim(-4,0)
ax.set_xlabel(r'$Z_1$', fontsize=18)
ax.legend()
ax = fig.add_subplot(212)
ax.hist(mc[:,1], normed=True, weights=mc[:,2], bins=nbins, histtype='step', label='MC');
ax.hist(data[:,1], normed=True, weights=data[:,2], bins=nbins, histtype='step', label='Data');
ax.set_xlim(-3,3)
ax.set_xlabel(r'$Z_2$', fontsize=18)
ax.legend()
```
